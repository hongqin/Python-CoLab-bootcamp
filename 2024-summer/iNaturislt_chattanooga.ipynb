```python
from pyinaturalist import get_observations
import pandas as pd
import folium
from folium.plugins import MarkerCluster

# Define the coordinates and radius for Chattanooga
latitude = 35.0456
longitude = -85.3097
radius = 10  # in kilometers

# Retrieve observations
observations = get_observations(
    lat=latitude,
    lng=longitude,
    radius=radius,
    per_page=100,  # Number of observations per page
    page=1,  # Page number
)

# Convert the observations to a DataFrame
obs_data = observations['results']
df = pd.DataFrame(obs_data)

# Extract latitude and longitude from the DataFrame
df['lat'] = df['location'].apply(lambda x: x[0])
df['lon'] = df['location'].apply(lambda x: x[1])

# Create a folium map centered on Chattanooga
map_center = [latitude, longitude]
m = folium.Map(location=map_center, zoom_start=12)

# Add a marker cluster to the map
marker_cluster = MarkerCluster().add_to(m)

# Add observations to the marker cluster
for idx, row in df.iterrows():
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=f"Species: {row['taxon']['name'] if row['taxon'] else 'Unknown'}",
    ).add_to(marker_cluster)

# Save the map to an HTML file
m.save('chattanooga_inaturalist_map.html')

# Display the map in a Jupyter Notebook (if running in Jupyter)
m
```

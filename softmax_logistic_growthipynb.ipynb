```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(logits):
    """Compute the softmax of a list of logits."""
    exp_logits = np.exp(logits)
    return exp_logits / np.sum(exp_logits, axis=0)

def logistic_growth(alpha, delta, t):
    """Compute the logits for logistic growth at time t."""
    return alpha + delta * t

# Parameters for the variants
alpha = np.array([0.5, 0.2, -0.1])  # Initial log-frequencies for the variants
delta = np.array([0.05, 0.03, 0.02])  # Growth rates for the variants
time_points = np.linspace(0, 100, 200)  # Time points to evaluate

# Compute the logits for each variant at each time point
logits = np.array([logistic_growth(a, d, time_points) for a, d in zip(alpha, delta)])

# Apply the softmax function to get the frequencies
frequencies = np.array([softmax(logit) for logit in logits.T]).T

# Plotting the results
plt.figure(figsize=(10, 6))
for i, freq in enumerate(frequencies):
    plt.plot(time_points, freq, label=f'Variant {i+1}')
plt.xlabel('Time')
plt.ylabel('Frequency')
plt.title('Logistic Growth Estimation for SARS-CoV-2 Variants')
plt.legend()
plt.show()
```
